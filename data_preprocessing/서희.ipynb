```python
# 1) Colab에서 CSV 파일 업로드하기
import pandas as pd

df = pd.read_csv('/content/labelled_professor_scores.csv')

```

```python
print(df.columns.tolist())
```

```python
# 1) category_cols 재정의 (공백 포함)
category_cols = ['소통 잘됨', '참여 유도', '학점 잘줌', '친근함', '강의 잘함']

# 2) 교수님별 카테고리 비율 계산
df_ratio = df.copy()
df_ratio[category_cols] = df_ratio[category_cols] \
    .div(df_ratio[category_cols].sum(axis=1), axis=0)

# 3) 0–100 점수로 환산 (소수점 첫째 자리까지)
df_scores = df_ratio.copy()
df_scores[category_cols] = (df_scores[category_cols] * 100).round(1)

# 4) 결과 출력
from IPython.display import display
display(df_scores[['professor'] + category_cols])

```

```python
df_scores.to_csv('professor_scores_seohee.csv', index=False)
```
