```python
import pandas as pd
import numpy as np
import re
import chardet
from konlpy.tag import Komoran
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumGothic')

# 불용어 리스트
stopwords = [
    '강의', '수업', '교수님', '설명', '과제', '시험', '평가', '자료',  # 너무 일반적인 단어
    '좋', '잘', '많', '어렵', '쉽', '공정', '교수', '미영', '양은', '샘', '때'  # 어스펙트 키워드 중 일반적인 단어
]

# Komoran 초기화
komoran = Komoran()

# CSV 파일 인코딩 감지
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# 데이터 로드
file_path = "/content/merged_reviews_by_professor.csv"
encoding = detect_encoding(file_path)
try:
    df = pd.read_csv(file_path, encoding=encoding)
except UnicodeDecodeError:
    for enc in ['utf-8', 'cp949', 'euc-kr']:
        try:
            df = pd.read_csv(file_path, encoding=enc)
            break
        except UnicodeDecodeError:
            continue
    else:
        raise Exception("모든 인코딩 시도 실패")

# 텍스트 정제 함수
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text[:500]
    return ''

df['review_clean'] = df['review'].apply(clean_text)

# 어스펙트 키워드 정의
aspect_keywords = {
    "소통_잘됨": ["소통", "질문", "피드백", "답변", "응답", "의견", "상호작용", "대화"],
    "참여_유도": ["토론", "참여", "발표", "활동", "인터랙티브", "그룹", "프로젝트", "활기"],
    "학점_잘줌": ["학점", "성적", "평가", "관대한", "공정", "채점", "성적 잘", "학점 좋아"],
    "친근함": ["친절", "유머", "따뜻", "배려", "친근", "편안", "스스럼없이", "좋은 성격"],
    "강의_잘함": ["강의", "설명", "이해", "딕션", "명확", "논리적", "쉽게", "강의 잘", "알기 쉬움"]
}

# 어스펙트 추출
def extract_aspects(review, aspect_keywords):
    aspects = set()
    if not review:
        return list(aspects)
    pos = komoran.pos(review)
    nouns = [word for word, tag in pos if tag in ['NNG', 'NNP']]
    for aspect, keywords in aspect_keywords.items():
        for keyword in keywords:
            if keyword in nouns or keyword in review:
                aspects.add(aspect)
    return list(aspects)

df['aspects'] = df['review_clean'].apply(lambda x: extract_aspects(x, aspect_keywords))

# 감정 분석 파이프라인
sentiment_pipeline = pipeline("sentiment-analysis", model="alsgyu/sentiment-analysis-fine-tuned-model")

def get_sentiment(review):
    if not review:
        return 'LABEL_1', 0.0
    try:
        result = sentiment_pipeline(review, truncation=True, max_length=512)[0]
        return result['label'], result['score']
    except:
        return 'LABEL_1', 0.0

df[['sentiment_label', 'sentiment_score']] = df['review_clean'].apply(lambda x: pd.Series(get_sentiment(x)))
df['numerical_sentiment'] = df.apply(lambda row:
    row['sentiment_score'] if row['sentiment_label'] == 'LABEL_2' else
    (-row['sentiment_score'] if row['sentiment_label'] == 'LABEL_0' else 0),
    axis=1)

# 어스펙트 + 감정 점수 집계
df_valid = df[df['aspects'].map(lambda x: len(x) > 0) & (df['numerical_sentiment'] != 0.0)]
df_exploded = df_valid.explode('aspects')
aggregated = df_exploded.groupby(['professor', 'aspects'])['numerical_sentiment'].mean().reset_index()
aggregated.rename(columns={'aspects': 'aspect', 'numerical_sentiment': 'score'}, inplace=True)

# 어스펙트 수 기준 교수 TOP 5
df['aspect_count'] = df['aspects'].apply(len)
aspect_counts_by_professor = df.groupby('professor')['aspect_count'].sum().reset_index()
aspect_counts_by_professor = aspect_counts_by_professor.sort_values(by='aspect_count', ascending=False)
print("어스펙트가 많은 교수님 Top 5:")
display(aspect_counts_by_professor.head(5))

# 감정 분포 출력
print("감정 레이블 분포:")
display(df['sentiment_label'].value_counts())

# SBERT 기반 유사 교수 추천
professor_reviews = df.groupby('professor')['review_clean'].apply(lambda x: ' '.join(x)).reset_index()
ko_model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
professor_embeddings = professor_reviews['review_clean'].apply(lambda x: ko_model.encode(x)).tolist()

def get_similar_professors_semantic(professor_name, top_n=5):
    try:
        prof_idx = professor_reviews[professor_reviews['professor'] == professor_name].index[0]
        prof_vector = professor_embeddings[prof_idx]
        similarities = cosine_similarity([prof_vector], professor_embeddings)[0]
        top_indices = similarities.argsort()[::-1][1:top_n+1]
        return professor_reviews.iloc[top_indices]['professor'].tolist()
    except:
        return ["해당 교수 없음"]

# TF-IDF 키워드 추출
def extract_tfidf_keywords(professor, df, top_n=10):
    prof_reviews = df.groupby('professor')['review_clean'].apply(lambda x: ' '.join(x.dropna())).reset_index()

    # 불용어를 Komoran 명사로 변환
    def preprocess_stopwords(stopwords):
        processed_stopwords = []
        for word in stopwords:
            pos = komoran.pos(word)
            nouns = [w for w, t in pos if t in ['NNG', 'NNP']]
            processed_stopwords.extend(nouns)
        return processed_stopwords

    processed_stopwords = preprocess_stopwords(stopwords)

    vectorizer = TfidfVectorizer(
        analyzer='word',
        tokenizer=lambda x: [word for word, tag in komoran.pos(x) if tag in ['NNG', 'NNP']],
        token_pattern=None,
        stop_words=processed_stopwords,
        max_features=100
    )
    tfidf_matrix = vectorizer.fit_transform(prof_reviews['review_clean'])
    feature_names = vectorizer.get_feature_names_out()
    prof_idx = prof_reviews[professor_reviews['professor'] == professor].index[0]
    tfidf_scores = tfidf_matrix[prof_idx].toarray().flatten()
    top_indices = tfidf_scores.argsort()[::-1][:top_n]
    keywords = [(feature_names[idx], tfidf_scores[idx]) for idx in top_indices if tfidf_scores[idx] > 0]
    return dict(keywords)

# 워드 클라우드
def generate_wordcloud(professor, df):
    keywords = extract_tfidf_keywords(professor, df)
    if not keywords:
        return None
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        font_path=font_path,
        relative_scaling=0.5,
        max_words=20
    ).generate_from_frequencies(keywords)
    return wordcloud

# 시각화 1: 어스펙트별 감정 점수 히트맵
pivot_df = aggregated.pivot(index='professor', columns='aspect', values='score')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', center=0, fmt='.2f',
            linewidths=0.5, linecolor='gray', cbar_kws={'label': '감정 점수'})
plt.title('교수님별 어스펙트 평균 감정 점수')
plt.xlabel('어스펙트')
plt.ylabel('교수님')
plt.tight_layout()
plt.show()

# 시각화 2: 감정 분포 바 차트
sentiment_counts = df.groupby('professor')['sentiment_label'].value_counts().unstack(fill_value=0)
sentiment_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['blue', 'gray', 'orange'])
plt.title('교수님별 감정 분포')
plt.xlabel('교수님')
plt.ylabel('리뷰 수')
plt.legend(['부정', '중립', '긍정'])
plt.tight_layout()
plt.show()

# 시각화 3: 전체 리뷰 워드 클라우드
all_reviews = ' '.join(df['review_clean'])
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    font_path=font_path,
    stopwords=stopwords
).generate(all_reviews)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('전체 리뷰 워드 클라우드')
plt.show()

# 시각화 4: 교수별 워드 클라우드
for prof in df['professor'].unique():
    wordcloud = generate_wordcloud(prof, df)
    if wordcloud:
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(f"{prof} 교수님 워드 클라우드")
        plt.tight_layout()
        plt.show()

# 추천 예시 출력
top_professor = aspect_counts_by_professor.iloc[0]['professor']
print(f"'{top_professor}' 교수님과 유사한 교수님 추천:")
print(get_similar_professors_semantic(top_professor))
```

```python
# 예시: 이미 위에서 만든 aggregated 데이터프레임을 활용
# aggregated: columns=['professor', 'aspect', 'score']

def recommend_top_professors_by_aspect(aggregated, top_n=1):
    aspect_list = aggregated['aspect'].unique()
    result = {}
    for aspect in aspect_list:
        sub = aggregated[aggregated['aspect'] == aspect]
        # score가 높은 순으로 정렬
        top_profs = sub.sort_values('score', ascending=False).head(top_n)
        # 결과를 {aspect: [(professor, score), ...]} 형식으로 저장
        result[aspect] = list(zip(top_profs['professor'], top_profs['score']))
    return result

# 사용 예시
recommendations = recommend_top_professors_by_aspect(aggregated, top_n=1)
for aspect, profs in recommendations.items():
    print(f"[{aspect}] 추천 교수:")
    for prof, score in profs:
        print(f"  {prof} (평균 감정 점수: {score:.2f})")
    print()
```

```python
# 사용자 지정 가중치
aspect_weights = {
    "소통_잘됨": 5.0,
    "참여_유도": 4.0,
    "학점_잘줌": 3.0,
    "친근함": 2.0,
    "강의_잘함": 1.0
}

# 교수별 총합 점수 계산
weighted_scores = aggregated.copy()
weighted_scores['weighted_score'] = weighted_scores.apply(
    lambda row: row['score'] * aspect_weights.get(row['aspect'], 1.0), axis=1
)

# 교수별 총합
prof_total_scores = weighted_scores.groupby('professor')['weighted_score'].sum().reset_index()
prof_total_scores = prof_total_scores.sort_values(by='weighted_score', ascending=False)

# 추천 결과 출력
print("📌 사용자 선호 기반 추천 교수 Top 5:")
display(prof_total_scores.head(5))
```
